```python
import re
```

ADDRESSES

```python
with open("addresses.txt", "r+", encoding="utf-8") as o:
  text = o.read()
```

```python
re.findall("\d+.+\n.+\n.+\d{5}", text)
```

DIRECT SPEECH

```python
with open("d_speech.txt", "r+", encoding="utf-8") as o:
  text = o.read()
```

```python
re.findall("— .+?—|— .+|«.+?—|—.+»", text)
```

SUBSTITUTION

```python
with open("ея.txt", "r+", encoding="utf-8") as o:
  text = o.read()
```

```python
re.sub("яз(ы|ы́)к", "шашл" + r"\1" + "к", text)
```
