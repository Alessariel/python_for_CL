**Задача:**

 извлечь и посчитать все n-граммы из слов из текта на русском языке.

1. sklearn

```python
bigram_vectorizer = CountVectorizer(ngram_range=(2,4),
                                    token_pattern=r'\b\w+\b', min_df=1)
# изменяемые паттерны: 
# token_pattern=r'здесь регулярка'
# ngram_range=(от,до)

analyze = bigram_vectorizer.build_analyzer()
```

```python
z = analyze(text) # сохраним результат в переменную

Counter(z).most_common() # посчитаем самые частотные n-граммы

# обратите внимание, что переменную z мы подаем в Counter()
```

```python
# запишем функцию, в которой используем функцию от nltk!

def ngrams_by_nltk(word, num): 
#создаем функцию, прописываем аргументы: строка + чему равно n в n-gram(биграммы, триграммы, и т.д.)

    word = word.lower() # предобработка: сводим к нижнему регистру
    
    word = re.sub("[!\"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~]?","",word)
    #убираем пунктуацию

    tokens = [token for token in word.split() if token != ""]
    # добавляем токены в список, если они не пустые. Дальше этот список будем подавать функции как аргумент

    output = list(ngrams(tokens, num)) 
    # используем встроенную ngrams, а ее результат делаем листом и сохраняем в переменную output

    return output
```

```python
ngrams_by_nltk(text,3) # 
```

```python
res3 = ngrams_by_nltk(text,3) # сохраним результат в переменную
```

```python
Counter(res3).most_common() # считаем самые частотные n-граммы
```
