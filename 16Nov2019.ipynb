1. sklearn

```python
bigram_vectorizer = CountVectorizer(ngram_range=(2,4),
                                    token_pattern=r'[A-Za-z]+', min_df=1) # используем импортированный модуль
# изменяемые параметры: 
# token_pattern=r'здесь регулярка', которая задает искомый паттерн
# ngram_range=(от,до) задает количество n-грамм (например, найти все от 2грамм до 4грамм)

analyze = bigram_vectorizer.build_analyzer() # применили функцию, сохранили в переменную
```

2. попробуем NLTK

```python
from nltk.util import ngrams
# ngrams - это встроенная функция в модуле nltk
```

```python
# здесь можно почитать подробнее, как она работает
? ngrams
```

```python
# запишем функцию, в которой используем функцию от nltk!

def ngrams_by_nltk(word, num): 
#создаем функцию, прописываем аргументы: строка + чему равно n в n-gram(биграммы, триграммы, и т.д.)

    word = word.lower() # предобработка: сводим к нижнему регистру
    
    word = re.sub("[!\"#$%&\'—()*+,-»«./:;<=>?@[\\]^_`{|}~]?","",word)
    #убираем пунктуацию

    tokens = re.findall("я\s\w+",word) # здесь задаем паттерн

    #
    # добавляем токены в список, если они не пустые. Дальше этот список будем подавать функции как аргумент

    output = list(ngrams(tokens, num)) 
    # используем встроенную ngrams, а ее результат делаем листом и сохраняем в переменную output

    return output
```

```python
ngrams_by_nltk(text,2)  
```

```python
res3 = ngrams_by_nltk(text,2) # сохраним результат в переменную
type(res3)
```

```python
Counter(res3).most_common() # считаем самые частотные n-граммы
```
